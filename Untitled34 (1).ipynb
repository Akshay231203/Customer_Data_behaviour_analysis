```python
import psycopg2
from psycopg2 import sql

user = "postgres"
password = "1234"
host = "localhost"
port = 5432
dbname = "customer_behavior"

try:
    con = psycopg2.connect(dbname="postgres", user=user, password=password, host=host, port=port)
    con.autocommit = True
    cur = con.cursor()
    cur.execute("SELECT 1 FROM pg_database WHERE datname=%s;", (dbname,))
    if cur.fetchone():
        print(f"Database '{dbname}' already exists.")
    else:
        cur.execute(sql.SQL('CREATE DATABASE {}').format(sql.Identifier(dbname)))
        print(f"Database '{dbname}' created.")
    cur.close()
    con.close()
except Exception as e:
    print("Error creating/checking database:", e)
```

```python
from sqlalchemy import create_engine, text
from urllib.parse import quote_plus

username = "postgres"
password = "1234"
host = "localhost"
port = 5432
database = "customer_behavior"   # now exists!

# encode password in case it has special characters
password_quoted = quote_plus(password)

# build SQLAlchemy connection URL
url = f"postgresql+psycopg2://{username}:{password_quoted}@{host}:{port}/{database}"

# create the engine
engine = create_engine(url, connect_args={"connect_timeout": 10})

# quick test
with engine.connect() as conn:
    version = conn.execute(text("SELECT version();")).scalar()
    print("✅ Connected to PostgreSQL — server version:", version)
```

```python
table_name = "customer"

df.to_sql(
    table_name,
    con=engine,
    if_exists="replace",   # or "append" if you want to add data
    index=False,
    method="multi",
    chunksize=1000
)

print(f"🎉 Data successfully loaded into table '{table_name}' in database '{database}'.")
```

```python
import pandas as pd

check_df = pd.read_sql_query(f"SELECT * FROM {table_name} LIMIT 5;", con=engine)
print(check_df)
```
